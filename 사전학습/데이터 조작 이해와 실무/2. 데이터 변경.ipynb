## 2-2. 인덱스 설정 및 리셋

```python
# 새로운 column 성별(gender) 생성
# 실행해보면 NaN 값으로 제대로 설정되지 않는 것을 확인할 수 있음
gender = pd.Series(['F', 'F', 'M', 'M', 'M'])
df['성별'] = gender
display(df)
```

```python
# df, gender Series 인덱스 확인
# 인덱스가 다르면 위와 같이 이용할 수 없다
print(df.index)
print(gender.index)
```

```python
# 인덱스를 어느 한 쪽에 맞춰준다면 결과와 같이 매칭할 수 있다
gender.index = df.index
df['성별'] = gender
display(df)
```

```python
# 현재 설정되어 있는 인덱스 reset
# 시게열 데이터를 다루거나 여러가지 이유로 인덱스를 재설정해야하는 경우가 종종 있다고 한다.
# 기존의 인덱스는 데이터프레임의 column으로 들어온다
df.reset_index()
```

```python
# display(df)
# 위 메소드 실행 후 데이터프레임을 다시 출력해보면 인덱스 리셋이 적용되지 않음을 확인할 수 있다
# 아직 원본에 대한 저장이 이뤄진 게 아니다
df.reset_index(inplace=True)
# 이렇게 inplace=True 옵션을 줌으로써 원본에 변경 사항을 저장해줄 수 있다
```

```python
# 필요한 라이브러리 로딩
import pandas as pd
import json      # json을 다루기 위한 라이브러리
import requests  # 서버에 request를 보내기 위한 라이브러리
```

```python
# Open API를 호출하기 위한 정보 입력
key = 'VXRWSv8jBsGyPRwRJebZ9dbGog5HRowmMOd%2BP9pizFvveUVoc6sb2hMcjhiJMVatkxB1fW7A2pdTGrVKfpsmrg%3D%3D'
# 필수적으로 적어야 하는 요청 변수들
pageNo = 0      # 페이지 번호
numOfRows = 100 # 한 페이지에 보이는 결과 개수

# 전체 요청 url 입력: 주소?serviceKey=키&pageNo=페이지번호&numOfRows=출력개수&type=json
url = 'http://api.data.go.kr/openapi/tn_pubr_public_unmanned_traffic_camera_api?serviceKey={key}&pageNo={pageNo}&numOfRows={numOfRows}&type=json'

response = requests.get(url) # requests 라이브러리를 활용해 API요청

# 한글 출력 시 encoding 필요
response.encoding = 'utf-8'
print(response.text)
```

```python
# 결과 json을 dictionary로 변환
resulting_dict = json.loads(response.text) # json을 dict 타입으로 변환
#print(resulting_dict['header'])
resulting_dict.keys()
```

```python
# dictionary의 key 값으로 response가 나왔는데, 원본 json은 보다시피 중첩 데이터임
resulting_dict['response'].keys()
```

```python
# 중첩 dict의 key값을 아래와 같이 계속해서 확인할 수 있음
resulting_dict['response']['header'].keys()
```

```python
# 첫 번째 아이템 호출
resulting_dict['resposne']['body']['items'][0]
```

```python
# 주소, 카메라 위치, 위도, 경도 정보만 출력
address_list = []
location_list = []
latitude_list = []
longitude_list = []

# dictionary 속에 데이터 정보 입력
for data in resulting_dict['response']['body']['items']:
    address_list.append(data['rdnmadr'])
    location_list.append(data['itlpc'])
    latitude_list.append(data['latitude'])
    longitude_list.append(data['longitude'])
    
# 불러온 정보 리스트 확인
print(address_list[:10])
print(location_list[:10])
print(latitude_list[:10])
print(longitude_list[:10])
```

```python
# list를 key에 매칭시켜 data frame으로 변환
df = pd.DataFram({
    'address': address_list,
    'location': location_list,
    'latitude': latitude_list,
    'longitude': longitude_list
})

display(df)
```
