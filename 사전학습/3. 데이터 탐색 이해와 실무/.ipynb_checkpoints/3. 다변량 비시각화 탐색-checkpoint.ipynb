# 3. 다변량 비시각화 탐색
<br>

### 다변량 비시각화의 종류
![image](https://github.com/domo7304/domo7304/assets/69956503/ff7e7c66-d82f-4639-a54f-5b39fd996d01)

<img width="500" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/3f6b132e-247b-4f40-aa78-cbda8afa2c03">

1. **교차표 : 범주형-범주형 변수 조합 간 연관 관계 파악**
    - ex) 소득 범주와 차량 보유 사이의 상관 관계 비교 (소득 범주에 따른 차량 보유 여부 확인)
    - 이 때 거주 지역 정보를 추가하여 소득 범주, 거주 지역에 따른 차량 보유 여부 등으로 데이터 간 연관 관계를 파악해 볼 수 있다.
2. **범주별 요약 통계량 : 범주형-연속형 변수 조합 간 범주별 대표 수치 비교**
    - ex) 소득 범주와 연령 사이의 상관 관계 비교 (소득 범주에 따른 평균 연령대 확인)
3. **상관계수 : 연속형 - 연속형 변수 조합 간 관계성 강도 파악**
    - ex) 연령, 연소득, 가구 구성원 사이의 상관계수 도출
    - 다른 변수 간 관계성에 비해 연령-연소득의 상관관계가 높음을 확인할 수 있다.
<br><br>
### 높은 상관계수
- 독립변수 간에 **강한 상관관계**가 발생할 경우 **다중공선성**이 발생할 수 있다.
- 다중공선성이란 **독립 변수의 일부가 다른 독립 변수의 조합으로 표현될 수 있는 경우**를 말한다.
- 이러한 경우 '독립 변수 간의 관계는 독립적'이라는 **회귀분석 가정**에 위배된다.
- 현실적으로 모든 변수들이 독립적일 수는 없지만 모델링 사전 단계 내에서 상관 분석 결과를 고려할 필요가 있다.
<br><br>


1. **교차표** : 범주형-범주형 다변량 비시각화
<img width="142" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/2f2a2d11-cef7-457d-b9b3-c4690e4550e5">

    - '주택 가격'과 '찰스강 경계 위치 여부' 사이의 상관 관계를 비교
    - 연속형 변수인 주택 가격을 구간으로 나누어 범주화(이산화)
    - pd.crosstab() 메서드를 활용해 강 경계 위치와 주택 가격 범주 사이의 연관관계 시각화
    - 강 경계 위치 여부와 주택 가격의 상관 관계가 크지 않다는 것을 파악할 수 있음
    ---
    <img width="277" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/1519bed5-8d79-4cfe-a425-1afb2ad33cbf">

    - 다른 기준을 확인해보기 위해 '비소매 사업 지역'을 확인. 해당 연속형 변수를 범주화
    - 도출 결과, '비소매 사업 지역' 비율이 평균 대비 높을수록 주택가격이 평균 대비 낮은 지역이 많음.
    - 비소매 사업 지역 비율이 평균 대비 낮은 경우 주택 가격은 큰 차이가 없음을 확인할 수 있음
    ---
    <img width="270" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/2e5bb4d8-e8fb-42a8-b6b2-47a80ac53e0e">

    - '고속도로 접근성 지수'를 확인. 해당 변수 역시 LOW, HIGH로 범주화
    - 고속도로 접근성 지수가 높은 경우 주택 가격이 평균 대비 낮은 지역이 많음
    - 접근성 지수가 낮은 경우 비등한 관계를 보임
    ---
    <img width="328" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/5e2ed4f7-7a3e-4fa6-968a-67dc7dcf18d7">

    - '비소매 사업 지역', '고속도로 접근성 지수' 를 합쳐 교차표 생성
    - 여전히 LOW, LOW 구분에서는 연관관계가 잘 파악되지 않음을 확인할 수 있다.
    - 위와 같은 경우에 대해서는 범주형 변환 및 분석에 활용할 때 더 세밀하게 범주 범위를 조정할 필요가 있음

    - 'RAD_LOW', 'INDUS_G3' 을 제외하고는 구분이 조금 더 명확해짐

<br>

2. **범주별 요약 통계** : 범주형 - 연속형 변수 조합 간 대표 수치 도출
    - 위 과정과 유사하게 데이터를 파악하는 작업 진행, 생략
<br><br>

3. **상관관계** : 연속형 - 연속형 변수 조합 간 관계성 강도 파악
    - 상관계수 도출 방안
        - corrwith : 한 column과 다른 모든 column들의 상관관계
        - corr : 모든 column 간의 상관관계
    - 상관관계는 타겟 변수와 독립 변수 간의 관계를 보는 경우가 많다. 
    - 타겟과 밀접한 연관을 지닌 독립 변수가 무엇인지, 얼마나 밀접한지 확인하기 위함
    <img width="85" alt="image" src="https://github.com/domo7304/domo7304/assets/69956503/f1f5ef2a-0d9f-4f32-8c02-9198ac021e1e">
    
    - RM : 주택 1가구 당 평균 방의 개수
    - LSTAT : 하위 계층의 비율
    - 주택 가격이 높을수록 평균 방의 개수는 많고, 주택 가격이 낮을수록 하위 계층의 비율이 높다는 상관관계를 확인할 수 있다.
    - 이 때 매우 강한 상관관계를 보일 경우, Regression 분석 알고리즘을 활용할 때 다중공선성 문제에 직면할 수 있다.
    - 위 사항을 해결하기 위해 **PCA를 수행하는 등 새로운 특징을 생성하는 방안 고려 가능**
