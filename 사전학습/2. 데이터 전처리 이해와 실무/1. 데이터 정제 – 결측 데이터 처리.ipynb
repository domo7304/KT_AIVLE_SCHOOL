### 실습 내용 요약
- 예제 데이터 내 결측치를 생성하고, 결측치제거 및 결측치 대체 방안에 대한 전반적 실습<br><br>

### 데이터 전처리 과정 내 주의사항 ###
- 데이터 전처리 과정 진행 시에는 원본 데이터 copy 필수
- Python은 copy함수를 사용하지 않으면 원본 데이터의 값을 변경시킨다.
- 만일 전처리 내역이 변경되는 경우 데이터 로딩부터 모든 과정을 다시 시작해야함
- 따라서 원본 데이터와 전처리 과정을 진행할 데이터를 구분해 작업 수행


```python
import numpy as np
import pandas as pd
```

```python
# 데이터 로딩 및 개요 확인
cancer = pd.read_csv("./data/wdbc.data", header=None)
cancer
```

```python
# 데이터 column명 지정
cancer.columns = ["id","diagnosis","radius_mean","texture_mean","perimeter_mean",
                  "area_mean","smoothness_mean","compactness_mean","concavity_mean",
                  "concave points_mean","symmetry_mean","fractal_dimension_mean","radius_se",
                  "texture_se","perimeter_se","area_se","smoothness_se","compactness_se",
                  "concavity_se","concave points_se","symmetry_se","fractal_dimension_se",
                  "radius_worst","texture_worst","perimeter_worst","area_worst",
                  "smoothness_worst","compactness_worst","concavity_worst",
                  "concave points_worst","symmetry_worst","fractal_dimension_worst"]
# ID를 index화
cancer = cancer.set_index('id')
cancer
```

```python
# 데이터 복사
cancer_data = cancer.copy()

# 데이터 내 결측치 생성
# 실습을 위한 일부 데이터 선택
cancer_data = cancer_data[0:30]
cancer_data = cancer_data[['diagnosis', 'radius_mean', 'texture_mean', 
                           'perimeter_mean', 'area_mean']]
# 결측치 생성
# 6개 record에 대해 결측치 생성
cancer_data.iloc[2, :] = np.nan
cancer_data.iloc[5, 0] = np.nan
cancer_data.iloc[10, [3, 4]] = np.nan
cancer_data.iloc[12, 2:4] = np.nan
cancer_data.iloc[15, [0, 3]] = np.nan
cancer_data.iloc[24, 4] = np.nan

cancer_data
```

## 1-1. 결측치 제거하기

### 1) listwise deletion

```python
# 데이터 개요
cancer_data.info()
```

```python
# 30개 record 중, 6개 record에서 결측치 존재
# listwise deletion 수행
cancer_copy = cancer_data.copy()
cancer_copy = cancer_copy.dropna()

# 데이터 요약 : 총 30개의 record 중, 하나의 결측치라도 보유한 6개 record 삭제
print(cancer_copy.info())

# 데이터 차원 확인
print("데이터 차원: ", np.shape(cancer_copy))
```

```python
cancer_copy
```

### 2) pairwise deletion

```python
# 30개 record 중 1개 record에 대해서만 모든 변수 내 결측치 존재
# pairwise deletion 수행
# 모든 열이 결측치인 행만 삭제
cancer_copy = cancer_data.copy()
cancer_copy = cancer_copy.dropna(how='all')

# 데이터 요약: 총 30개 record 중 1개 record 삭제
print(cancer_copy.info())

# 데이터 차원 확인
print("데이터 차원: ", np.shape(cancer_copy))
```

```python
cancer_copy
```

## 1-2. 결측치 대체하기

```python
# 결측치 데이터 확인
cancer_copy = cancer_data.copy()
cancer_copy
```

#### 1) 일정 값 대체

```python
# 일정 값 대체 수행
# diagnosis 컬럼 내 결측치는 C라는 범주형 값으로 일괄 대체
cancer_copy['diagnosis'] = cancer_copy['diagnosis'].fillna('C')
cancer_copy.head(10)
```

```python
# 수치형 column인 radius_mean 컬럼 내 결측치는 65라는 수치의 일정 값으로 대체
cancer_copy['radius_mean'] = cancer_copy['radius_mean'].fillna(65)
cancer_copy.head(10)
```

```python
# 데이터 개요 확인
cancer_copy.info()
# 대체를 수행한 diagnosis, raduis_mean 컬럼과 그 이외 컬럼별 결측치 여부 확인
```

```python
# 특정 값 지정이 아니라 여러 다양한 값으로 대체 가능 (평균, 중앙, 최소, 최대 등)
# texture_mean 컬럼 내 결측치를 texture_mean 평균값으로 대체
# replace를 통해서로 fillna와 동일한 결과를 얻을 수 있다.
# cancer_copy['texture_mean'] = cancer_copy['texture_mean'].fillna(cancer_copy['texture_mean'].mean())
cancer_copy['texture_mean'] = cancer_copy['texture_mean'].replace(np.nan, cancer_copy['texture_mean'].mean())

# 대체된 값과 texture+mean 컬럼의 평균값 비교
# 전부 NaN이었던 3번째 record, id 84300903 확인해보기
print(cancer_copy['texture_mean'].mean())
cancer_copy.head(10)
```

#### 2) 선형 값 대체

```python
# 데이터의 선형 관계를 기반으로 대체
cancer_copy = cancer_data.copy()
cancer_copy.head()
```

```python
# 선형보간법
cancer_copy = cancer_copy.interpolate()
cancer_copy.head()
```

```python
# 선형 값 대체 확인
# radius_mean 컬럼 id 84300903을 살펴보면 평균값으로 대체되었음을 확인할 수 있다
# 선형 값 대체의 경우 데이터의 연속성을 기반으로 연산되므로 신중히 사용해야 한다
print((cancer_data.iloc[1, 1] + cancer_data.iloc[3, 1]) / 2)
```
