# 5. 데이터 축소 – 특징 선택

### 특징 선택
1. Filter 방식 : 특징들에 대한 통게적 기법을 기반으로 점수와 순위를 부여해 특징을 선택, 연산 속도가 비교적 빨라 Wrapper 방식과 같이 연산 비용이 많이 필요한 작업 이전에 거치기도 한다.
    - 카이제곱 필터(Chi-square filter) : 범주형 변수 간의 유의미성을 도출하기 위한 통계적 방안
    - 상관관계 필터(Correlation filter) : 연속적인 변수 간의 유의미성을 도출하기 위한 통계적 방안
<br><br>
2. Wrapper 방식 : 원본 데이터 내 변수 간 조합을 탐색해 특징 선택
    - 원본 데이터셋 내 변수들의 다양한 조합을 모델에 적용
    - 정확도 측면에서 가장 좋은 성능을 보이는 최적의 subset을 도출하는 방법론
    - 대표적 방식으로 재귀적 특성 제거(Recursive Feature Elimination) 존재, 전체 데이터셋에서 유의미하지 않은 subset을 제거해나가는 방식
<br><br>
3. Embedded 방식 : 모델을 학습하여, 정확도에 기여하는 특징을 선택하는 방안
    - 알고리즘 내 자체 내장 함수로 특징을 선택하는 방식, 모델 성능에 기여할 수 있는 특징을 도출
    - 모든 조합을 고려하고 결과를 도출하는 Wrapper와 달리 학습과정에서 최적화된 변수를 선택
    - 트리 계열 모델 기반의 특징 선택이 대표적 (랜덤포레스트 기반 Feature Inportance 기반) 
<br><br>
4. 보루타 알고리즘(Boruta Algorithm) : 랜덤포레스트 모형 기반의 알고리즘
    - 기존 데이터를 복제해 임의로 정한 랜덤(shadow)변수를 생성 후 데이터와 결합해, shadow보다 중요도가 낮을 경우 중요치 않는 변수라 판단하여 제거
    - 예측 성능과 강하게 연관을 지닌 기존 특징들이라면 랜덤하게 생성된 변수보다 그 주요도는 높을 것이기 때문에, shadow 변수보다 중요도가 낮은 변수들은 중요치 않다 판단하여 제거
    
<br><br>

## 5-1. Wrapper 기반 특징 선택

- 원 데이터 내 변수 간 조합을 탐색해 특징을 선택하는 방안

```python
import numpy as np
import pandas as pd
```

```python
# Breast Cancer 데이터 활용, 데이터 로딩
cancer = pd.read_csv("./data/wdbc.data", header=None)

# 데이터 컬럼명 지정
cancer.columns = ["id","diagnosis","radius_mean","texture_mean","perimeter_mean",
                  "area_mean","smoothness_mean","compactness_mean","concavity_mean",
                  "concave points_mean","symmetry_mean","fractal_dimension_mean","radius_se",
                  "texture_se","perimeter_se","area_se","smoothness_se","compactness_se",
                  "concavity_se","concave points_se","symmetry_se","fractal_dimension_se",
                  "radius_worst","texture_worst","perimeter_worst","area_worst",
                  "smoothness_worst","compactness_worst","concavity_worst",
                  "concave points_worst","symmetry_worst","fractal_dimension_worst"]

# ID를 Index화
cancer = cancer.set_index('id')
cancer
```

```python
# cancer Data Frame을 훈련, 테스트를 위한 데이터셋으로 나누기
# 특별히 지정하지 않으면 3:1 로 훈련:테스트 데이터가 나눠진다
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(cancer.drop(['diagnosis'], axis=1), 
                                                    cancer[['diagnosis']], random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 데이터가 지닌 특징을 보기 위해 훈련데이터와 테스트데이터를 merge
train_df = pd.merge(X_train, y_train, left_index=True, right_index=True, how='inner')
train_df
```

```python
# 각 변수별 시각화, 즉 양성, 음성별 데이터가 어떻게 분포되어 있는지를 시각화 실시
# 히스토그램을 사용
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(10, 3, figsize=(20, 40))
axes = axes.flatten()

for i in range(30):
    sns.histplot(data=train_df, x=train_df.columns[i], hue='diagnosis', 
                 multiple='layer', ax=axes[i])
    
plt.show()

# 실행결과 smoothness_mean 같은 인자를 보면, 해당 데이터가 양성과 음성을 뚜렷하게 구분한다고 보기 힘들다.
```

### 1) RFE (Recursive Feature Elimination) 기반 주요 특징 선택

```python
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
```

### SVM 기반의 RFE 방식
**Guyon, isabelle, et al, "Gene selection for cancer classification using support vector machines." Machine learning 46.1-2 (2002): 389~422.**
- 위 논문은 SVM 알고리즘 기반의 RFE를 적용해 대장암과 관련된 유전자 선별을 진행한 연구
- 4개의 유전자를 최종적으로 선택해 classification accuracy를 86%에서 98%로 개선했다고 결과 발표

```python
# SVM 기반 RFE 수행
from sklearn.preprocessing import StandardScaler
# 스케일에 민감한 SVM의 특징에 따라 변수의 scaling을 따로 진행

scaler = StandardScaler()
scaler.fit(X_train)
scale_X_train = scaler.transform(X_train)
```

```python
# RFE를 적용한 모델에 SVM 지정
estimator_mdl = SVC(kernel='linear')

# SVM 학습 기반의 RFE 실행 및 최종적으로 남길 유의미한 특징 개수 지정 : 5개
svm_rfe = RFE(estimator = estimator_mdl, n_features_to_select=5)

svm_rfe_rst = svm_rfe.fit(scale_X_train, y_train.values.ravel())
svm_rfe_rst.ranking_
# 1로 나온 변수가 최종 유의미한 특징으로 도출된 column
```

```python
# 도출한 특징 조합으로 test 진행
prediction = pd.DataFrame(svm_rfe.predict(scale_X_test), columns=['pred_rst'])
prediction
```

```python
# 예측의 성능평가 진행
# 타겟 클래스 데이터를 수치형으로 변경
y_test['diagnosis'] = y_test['diagnosis'].replace('M', 1)
y_test['diagnosis'] = y_test['diagnosis'].replace('B', 0)
prediction['pred_rst'] = prediction['pred_rst'].replace('M', 1)
prediction['pred_rst'] = prediction['pred_rst'].replace('B', 0)
```

```python
# 결과 확인
from sklearn.metrics import accuracy_score
print('accuracy: ', round(accuracy_score(y_test['diagnosis'], prediction['pred_rst']), 5))

# auc
from sklearn.metrics import roc_auc_score
print('auc: ', round(roc_auc_score(y_test['diagnosis'], prediction['pred_rst']), 5))
```

```python
# 동일한 SVM RFE 방식을 수행하되, 10개 특징을 남기는 방법으로 실행
estimator_mdl = SVC(kernel='linear')
# SVM 학습 기반의 RFE 실행 및 최종적으로 남길 유의미한 개수 지정 : 10개
re_svm_rfe = RFE(estimator = estimator_mdl, n_features_to_select=10)
re_svm_rfe_rst = re_svm_rfe.fit(scale_X_train, y_train.values.ravel())

re_prediction = pd.DataFrame(re_svm_rfe.predict(scale_X_test), columns=['pred_rst'])
re_prediction['pred_rst'] = re_prediction['pred_rst'].replace('M', 1)
re_prediction['pred_rst'] = re_prediction['pred_rst'].replace('B', 0)

# 결과 확인
print('accuracy: ', round(accuracy_score(y_test['diagnosis'], re_prediction['pred_rst']), 5))

# auc
print('auc: ', round(roc_auc_score(y_test['diagnosis'], re_prediction['pred_rst']), 5))

# 확인 결과 10개 특징을 남기는 것이 5개 특징을 남기는 것보다 성능이 다소 낮음을 확인
```

```python
# 최종 선택된 5개의 특징 column이 어떤 것인지 확인
chk_var=[]
for name, svm_rfe_rst.ranking_ in zip(X_train.columns, svm_rfe_rst.ranking_): # 같이 묶어서 호출
    lst_chk= [name, svm_rfe_rst.ranking_]
    chk_var.append(lst_chk)
    
chk_svm_rfe = pd.DataFrame(chk_var, columns=['feature_names', 'svm_rfe_feature'])
chk_svm_rfe = chk_svm_rfe[chk_svm_rfe['svm_rfe_feature']==1]
chk_svm_rfe
```

### SVM-RFE 기반 판단
- 시각화 기반으로 유의할 것이라고 예상했던 특징 중 3개 특징이 포함
    - SVM이외에도 decision tree, logistic regression 등 알고리즘을 자유롭게 활용해 RFE를 할 수 있다.
    - RFE 내에서 사용된 모델이 특징 선택과 예측 문제 성능을 결정하는데 중요한 차이를 만들 수 있음

## 5-2. Embedded 기반 특징 선택

- 모델 학습 성능에 기여하는 특징을 선택하는 방안

```python
# Random Forest 기반 특징 선택
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
# RF 모형 생성
# Random Forest의 경우 나무(n_estimators)가 많아질수록 성능이 좋아질 수 있다고 기대는 할 수 있지만
# 무조건 향상된다 보장할 수는 없으며, 연산량이 많아진다는 것을 생각하고 있어야 한다.
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=1))
embeded_rf_selector.fit(scale_X_train, y_train.values.ravel())
```

```python
# RF 기반 Embedded 특징 선택 결과
embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X_train.loc[:, embeded_rf_support].columns.tolist()
print(str(len(embeded_rf_feature)), 'selected features')
```

```python
# 선택된 9개의 특징
embeded_rf_feature
```

### Embedded (Random Forest) 기반 판단
- 시각화 기반으로 유의할 것이라고 예상했던 특징 중 8개 특징이 포함

## 5-3. Boruta Algorithm

- Boruta Algorithm은 Random Forest 모형 기반의 특징 선택 알고리즘
- 기존 데이터를 복제해 임의의 변수 (Shadow)를 생성하고, 이보다 낮은 중요도를 지닌 특징을 제외함

```python
from boruta import BorutaPy
```

```python
# boruta 알고리즘은 Random Forest 모형 기반이므로 먼저 RF 모형 설정
rf = RandomForestClassifier(random_state=1)

# boruta 기반 특징 선택
boruta_selector = BorutaPy(rf, n_estimators='auto', random_state=1)

# 'auto' 인 경우는 자동으로 데이터셋 사이즈를 고려해 자동 설정

boruta_selector.fit(scale_X_train, y_train.values.ravel())
boruta_selector.ranking_
```

```python
# 최종 선택된 column이 어떤 column인지 확인
# 꽤 많은 특징들이 선택된다.
chk_var_boruta=[]
for name, boruta_selector.ranking_ in zip(X_train.columns, boruta_selector.ranking_):
    lst_chk = [name, boruta_selector.ranking_]
    chk_var_boruta.append(lst_chk)
    
chk_boruta = pd.DataFrame(chk_var_boruta, columns=['feature_names', 'boruta_feature'])
chk_boruta = chk_boruta[chk_boruta['boruta_feature'] == 1]
chk_boruta
```

```python
# boruta 알고리즘을 통해 23개의 특징 선택
print(np.shape(chk_boruta))
```

```python
# 최종 제외된 특징들 리스트 확인
non_chk_boruta = pd.DataFrame(chk_var_boruta.columns=['feature_names', 'boruta_feature'])
non_chk_boruta = non_chk_boruta[non_chk_boruta['boruta_feature'] != 1]
non_chk_boruta
```

### Boruta Algorithm 기반 판단
- 시각화 기반으로 유의할 것이라고 예상했던 특징 중 6개 특징이 포함
