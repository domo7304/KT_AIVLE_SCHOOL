# 5. 시계열 데이터 기초

## 5-1. pd.to_datetime

```python
import pandas as pd # pandas의 DatatimeIndex를 사용할 예정
import numpy as np # random seed 및 난수 생성
```

```python
# 아래와 같이 임의의 날짜 문자열을 생성
# yyyy/mm/dd 뿐만 아니라 마침표, 콤마, 공백 등 다양한 포맷에서 to_datetime 함수를 이용할 수 있다
date = ["2020/01/01", "2020.02.01.", "2020,03,01", "2020 04 01"]
date_idx = pd.to_datetime(date) # 날짜 문자열 -> Datetimeindex

print(date_idx)
```

```python
np.random.seed(1) # 고정된 난수를 얻기 위해 seed값 설정

# 3과 10사이의 정수로 구성된 4개의 원소를 가진 numpy 난수 배열 생성
random_int = np.random.randint(3, 10, size=4)
random_int
```

```python
# numpy를 배열 Series로 변환
series = pd.Series(random_int, index=date_idx)
series
```

## 5-2. pd.date_range

```python
# data_range : 시작일, 종료일을 파라미터로 설정
pd.date_range('2020-01-01', '2020-05-31') # 해당 기간 일별 데이터로 출력
```

```python
# 월별 데이터 출력
# freq 속성을 통해 범위 조정 가능
pd.date_range('2020-01-01', '2020-05-31', freq='M')
```

```python
pd.date_range('2020-01-01', '2020-05-31', freq='MS')
```

아래는 많이 사용하는 freq 속성들
- D: 일별(default)
- W: 주별(일요일기준)
- W-MON: 주별(월요일기준)
- M: 월별(월의 마지막 날 기준)
- MS: 월별(월의 첫째 날 기준)
- B: 주말을 제외한 평일 날짜들(토, 일 날짜 제외)

```python
# 종료일을 명시하지 않고 횟수를 명시할 수도 있다
pd.date_range(start="2020-01-01", periods=45)
```

```python
# periods & freq 명시
pd.date_range(start="2020-01-01", periods=45, freq='MS')
```

## 5-3. 시계열의 shift

```python
np.random.seed(1) # 고정된 난수만 사용하기 위해 seed 설정
arr = np.random.randn(12) # 가우시안 정규분포 확률을 따르는 난수 12개 생성

ts_idx = pd.date_range('2020-01-01', periods=12, freq='MS') # 2020년 월 초를 기준으로 12개 데이터

# 날짜를 index로 하고, arr를 값으로 갖는 임의의 series 생성
ts= pd.Series(arr, index=ts_idx)
print(ts)
```

```python
# 1개월 이동
ts.shift(1)
```

```python
# 4개월 이동
ts.shift(4)
```

```python
# 각 월의 첫 주 일요일을 기준으로 데이터 shift 진행, 인덱스 날짜가 변경된다.
ts.shift(1, freq='W')
```

## 5-4. diff(n): n만큼 이전 데이터와의 차분값 계산

```python
# 1개월치 차분값 연산
ts.diff(1) # 차분값 = 현재 - 1달 전
```

```python
# 3개월치 차분값 연산
ts.diff(3) # 연산 방식은 이런 방식 : ts[3] - ts[0] 
```

## 5-5. resample

```python
# resample? 월별 데이터를 일별 데이터로 바꾸거나, 그 반대의 경우 등에 사용
np.random.seed(1) # 항상 그렇듯 고정된 난수 출력을 위함

arr = np.random.randn(365) # 365개의 난수 생성
time_idx = pd.date_range('2021-01-01', periods=365, freq='D') # 365개의 DatetimeIndex 생성

ts = pd.Series(arr, index=time_idx)
print(ts)
```

#### 1) down-sampling

```python
# 365일 일별 데이터를 12달 월별 데이터로 down samping
# 월별 평균 주가 등 가볍게 생각해봐도 쓰일 곳이 많다.
ts.resample('M').mean() # 일별 데이터를 월별 데이터로, 월의 평균으로 대표되도록 변환
```

#### 2) up-sampling

```python
arr = np.random.randn(10) # 새로운 난수 10개 생성
time_idx = pd.date_range('2021-01-01', periods=10, freq='W') # 주별 데이터 10개 생성

ts = pd.Series(arr, index=time_idx)

print(ts)
```

```python
# forward filling 방식
ts.resample('D').ffill().head(21) # ffill: 각 기간의 첫 일을 참고해 결측값 보간
```

```python
# backward filling 방식
ts.resample('D').bfill().head(21) # bfill: 각 기간의 마지막 날을 참고해 결측값 보간
```

## 5-6. 시계열 데이터 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)

ts_data = np.random.randn(365, 3) # (365 x 3) 행렬의 난수 데이터를 생성
ts_idx = pd.date_range(start='2021-01-01', periods=365, freq='D') # 365개의 시계열 인덱스 생성

ts_df = pd.DataFrame(ts_data, index=ts_idx)
ts_df
```

```python
# column 이름 변경
ts_df.columns = ['X', 'Y', 'Z']
ts_df
```

```python
# 단순하게 .plot() 메소드를 호출해 시각화 가능
ts_df.plot() 
```

```python
ts_df.plot() 
plt.title('Time Series Plot of Random Numbers')
plt.show() # 그래프 출력 메소드
```

```python
# 누적합 메소드를 이용해 알아보기 쉽게 각 Series를 변경
ts_df = ts_df.cumsum()
ts_df.plot()
plt.title('Time Series Plot of Random Numbers (Cumulative)')
plt.show()
```

```python
# 가로 세로 크기 조정 및 x, y 축 이름 지정
ts_df.plot(figsize=(10, 3))
plt.title('Time Series Plot of Random Numbers')
plt.xlabel('Date')
plt.ylabel('Unit')
plt.show()
```
